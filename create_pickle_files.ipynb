```python
import librosa
import numpy as np
import os
import pickle
import sys
```

```python
def get_mfcc(signal):
    ss, sr = librosa.load(signal)
    return librosa.feature.mfcc(y=ss, sr=sr)
```

```python
# List all the directories
libri_dir = './LibriSpeech100/'
train_dir = libri_dir + 'train/train-clean-100/'
dev_dir = libri_dir + 'dev/dev-clean/'
test_dir = libri_dir + 'test/test-clean/'
```

```python
# create list of all folders
train_list = sorted(os.listdir(train_dir))
dev_list = os.listdir(dev_dir)
test_list = os.listdir(test_dir)
```

### Create Pickle Dump for Every SubFolder in Train Directory

Selects every flac file in the sub folder. Gets MFCC of that flac file and adds it to the main dictionary which in the end is stored as the pickle file. For the dictionary keys are the filename (without extension) and values are the MFCC spectrograms of the flac files

```python
for folder in train_list:
    print(folder)
    folder_list = os.listdir(train_dir+folder)
    for folder2 in folder_list:
        print(folder+'/'+folder2)
        file_list = os.listdir(train_dir+folder+'/'+folder2)
        pickle_dump = {}
        for file in file_list:
            file_name = file.split('.')
            if file_name[-1] != 'txt' and file_name[-1] != 'pkl':
                pickle_dump[file_name[0]] = get_mfcc(train_dir+folder+'/'+folder2+'/'+file)
        
        file_save = train_dir+folder+'/'+folder2+'/' + folder+'-'+folder2 + '.pkl'      
        print(file_save)
        with open(file_save, 'wb') as pickle_file:
            pickle.dump(pickle_dump, pickle_file)
```
